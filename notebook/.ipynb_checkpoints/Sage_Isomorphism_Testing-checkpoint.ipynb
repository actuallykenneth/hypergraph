```python
M1 = Matrix([
        (1,1,0,0,0,0),
        (1,0,1,0,1,0),
        (0,0,0,0,1,1),
        (0,0,1,1,0,1)
    ])

M2 = Matrix([
        (1,1,0,0,0,0),
        (2,0,2,0,2,0),
        (0,0,0,0,2,2),
        (0,0,1,1,0,1)
    ])

M3 = Matrix([
        (1,1,1,0,0,0),
        (0,0,2,2,0,0),
        (2,0,0,2,0,2),
        (0,0,0,0,1,1)
    ])

M3_v2 = Matrix([
        (1,1,1,0,0,0),
        (0,0,1,1,0,0),
        (1,0,0,1,0,1),
        (0,0,0,0,1,1)
    ])
```

```python
B1 = BipartiteGraph(M1)
B2 = BipartiteGraph(M2)
B3 = BipartiteGraph(M3)
B3_v2 = BipartiteGraph(M3_v2)
```

```python
B1.plot()
```

```python
B2.plot()
```

```python
B3.plot()
```

```python
B2.is_isomorphic(B3_v2)
```

```python
File = open('./uf50-01.clean.2')
firstLine = File.readline()
numRows = int(firstLine.split(" ")[0])
numCols = int(firstLine.split(" ")[1])

m = Matrix(nrows=numRows * 2, ncols=numCols+numRows)

for n,i in enumerate(File.readlines()):
    split_line = i.split(" ")
    print(n,i)
    for v in split_line:
        if int(v) < 0:
            m[(abs(int(v))-1 + (numRows)),n] = 1
        else:
            m[int(v) - 1,int(n)] = 1
            
for i in range(numRows):
    m[i,numCols + i] = 2
    m[numRows+i, numCols+i] = 2
```
